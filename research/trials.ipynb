```python
text_chunks = []
with pdfplumber.open('data/AIMI-3-55.pdf') as pdf:
    for page in pdf.pages:
        page_text = page.extract_text() or ""
        text_chunks.append(page_text)
    full_text = "\n".join(text_chunks)


match = re.search(r"(?im)^\s*references\s*$", full_text)

start_index = match.start()

# ---- 3) Slice from 'References' to the end ----
references_text = full_text[start_index:]

# Optionally, strip leading/trailing whitespace
print(references_text)
```
